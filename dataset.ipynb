```python
%%capture
from TTS.api import TTS
tts = TTS(model_name="tts_models/en/ljspeech/glow-tts", progress_bar=False)
```

```python
with open('sentences.txt') as f:
  i = 0
  while True:
    line = f.readline()
    if not line:
      break
    tts.tts_to_file(text=line, file_path=f"train_data//audio//output{i}.wav")
    i += 1
```
